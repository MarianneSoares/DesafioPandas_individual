```python
import pandas as pd
import numpy as np
```

```python
caminho = "sono.csv"
df = pd.read_csv(caminho)
df.head()
```

```python
#1
df = df.rename(columns={
    'ID': 'Identificador',
    'Ocupação': 'Profissão',
    'Categoria BMI': 'Categoria IMC',
    'Pressão sanguíneaaaa': 'Pressão Sanguínea'
})
df.head()
```

```python
#2
moda = lambda x: x.mode().iat[0]
agregacoes = {
    'Duração do sono': ['mean', moda, 'median']
    #
}

resultado = df.groupby('Profissão').agg(agregacoes)

resultado.columns = ['Média', 'Moda', 'Mediana']

resultado
```

```python
#3
eng_software = df[df['Profissão'] == 'Eng. de Software']

obesos_eng_software = eng_software[eng_software['Categoria IMC'] == 'Obesidade'].shape[0]

porcentagem_obesos = (obesos_eng_software / eng_software.shape[0]) * 100

print(f"Porcentagem de obesos entre engenheiros de software: {porcentagem_obesos:.2f}%")
```

```python
#4
media_sono_advogado_vendas = df[df['Profissão'].isin(['Advogado', 'Representante de vendas'])]['Duração do sono'].mean()

media_sono_outros = df[~df['Profissão'].isin(['Advogado', 'Representante de vendas'])]['Duração do sono'].mean()

dorme_menos_advogado_vendas = media_sono_advogado_vendas < media_sono_outros

if dorme_menos_advogado_vendas:
    print("advogar ou ser representante de vendas faz você dormir menos.")
else:
    print("advogar ou ser representante de vendas não faz você dormir menos.")
```

```python
#5

medias_sono_saude = df[df['Profissão'].isin(['Médico(a)', 'Enfermeiro(a)'])].groupby('Profissão')['Duração do sono'].mean()

profissao_maior_sono = medias_sono_saude.idxmax()
profissao_menor_sono = medias_sono_saude.idxmin()
menor_media_sono = medias_sono_saude.min()
maior_media_sono = medias_sono_saude.max()


print(f'A profissao com menor tempo de sono é o {profissao_menor_sono}, que tem uma media de {menor_media_sono:.2f}')
```

```python
#7
profissao_menos_frequente = df['Profissão'].value_counts().idxmin()


profissao_menos_frequente, df['Profissão'].value_counts()[profissao_menos_frequente]
```

```python
#8
def converter_pressao_sanguinea(pressao):
    sistolica, _ = pressao.split('/')
    return int(sistolica)


df['Pressão Sistólica'] = df['Pressão Sanguínea'].apply(converter_pressao_sanguinea)

media_pressao_por_genero = df.groupby('Gênero')['Pressão Sistólica'].mean()


genero_maior_pressao = media_pressao_por_genero.idxmax()
maior_media_pressao = media_pressao_por_genero.max()

genero_maior_pressao, maior_media_pressao
```

```python
#9
moda_duração_sono = df['Duração do sono'].mode().iat[0]
maioria_dorme_mais_de_8_horas = moda_duração_sono > 8

if maioria_dorme_mais_de_8_horas:
    print("A maioria dos participantes dorme mais de 8 horas por dia (moda = {})".format(moda_duração_sono))
else:
    print("A maioria dos participantes não dorme mais de 8 horas por dia (moda = {})".format(moda_duração_sono))
```

```python
#10
media_passos_acima_70 = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
media_passos_ate_70 = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()

mais_passos = media_passos_acima_70 > media_passos_ate_70

if mais_passos:
    print("Pessoas com frequência cardíaca acima de 70 dão mais passos do que pessoas com frequência cardíaca igual ou menor a 70.")
else:
    print("Pessoas com frequência cardíaca acima de 70 não dão mais passos do que pessoas com frequência cardíaca igual ou menor a 70.")
```
